```python
import sqlite3
```

```python
conn1 = sqlite3.connect(r"E:\\Dev_Environment\\equity-fundamental-engine\\data\\nse_equity_universe_bronze.db")
```

```python
cursor = sqlite3.Cursor(conn1)
```

```python
import re

def camel_to_snake(name):
    # Insert underscore before capital letters (except the first letter) and lowercase everything
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    s2 = re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1)
    return s2.lower()
```

```python
metrics = [row[0] for row in cursor.execute("""
    SELECT DISTINCT metric_name
    FROM raw_stock_cashflowstmt
    WHERE metric_name IN (
            'OperatingCashFlow',
            'FreeCashFlow',
            'CashFlowFromContinuingOperatingActivities',
            'OperatingCashFlow',
            'CapitalExpenditure',
            'CapitalExpenditureReported',
            'PurchaseOfPPE',
            'SaleOfPPE',
            'CommonStockDividendPaid',
            'PreferredStockDividendPaid',
            'CashDividendsPaid',
            'FinancingCashFlow',
            'RepurchaseOfCapitalStock',
            'IssuanceOfDebt',
            'IssuanceOfCapitalStock',
            'RepaymentOfDebt',
            'RepurchaseOfCapitalStock',
            'DividendsPaid'
        )
""")]
columns_sql = ",\n".join([f"MAX(CASE WHEN metric_name = '{m}' THEN value END) AS '{camel_to_snake(m)}'" for m in metrics])

query = f"""
SELECT
    symbol,
    fiscal_year,
    fiscal_date,
    {columns_sql}
FROM raw_stock_cashflowstmt
GROUP BY symbol, fiscal_year, fiscal_date;
"""
```
