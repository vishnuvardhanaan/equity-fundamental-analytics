```python
import pandas as pd
import sqlite3
```

```python
conn1 = sqlite3.connect(r"E:\\Dev_Environment\\equity-fundamental-engine\\data\\nse_equity_universe_bronze.db")
conn2 = sqlite3.connect(r"E:\\Dev_Environment\\equity-fundamental-analytics\\data\\nse_equity_universe_silver.db")
```

```python
with open(r"E:\\Dev_Environment\\equity-fundamental-analytics\\raw_equity_universe_clean.sql") as f:
    query = f.read()

df_reu = pd.read_sql(query, conn1)
```

```python
df_reu["clean_ingested_ts"] = pd.Timestamp.now(tz="UTC").tz_localize(None)
df_reu["raw_ingested_ts"] = pd.to_datetime(df_reu["raw_ingested_ts"])

for col in df_reu.select_dtypes(include="object").columns:
    df_reu[col] = df_reu[col].astype('string')
```

```python
df_reu.to_sql("clean_equity_universe", conn2, if_exists="replace", index=False)
```
