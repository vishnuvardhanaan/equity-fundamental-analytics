```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect(r"sandbox/dim_metrics.db")
conn2 = sqlite3.connect(r"data\gold\nse_equity_universe_gold.db")
```

```python
df_macro = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_macro")
df_macroweight = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_macroweight")
df_sector = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_sector")
df_sectorweight = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_sectorweight")
df_metrics = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_metrics")
df_absolute = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_absolute")
df_relative = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_relative")
df_trend = pd.read_excel(r"sandbox\dim_metrics.xlsx",sheet_name="dim_trend")
```

```python
df_macro.to_sql("dim_macro", conn, index=False, if_exists="replace")
df_sector.to_sql("dim_sector", conn, index=False, if_exists="replace")
df_metrics.to_sql("dim_metrics", conn, index=False, if_exists="replace")
df_absolute.to_sql("dim_absolute", conn, index=False, if_exists="replace")
df_relative.to_sql("dim_relative", conn, index=False, if_exists="replace")
df_trend.to_sql("dim_trend", conn, index=False, if_exists="replace")
```

```python
df_metrics.to_sql("dim_metrics", conn2, index=False, if_exists="replace")
df_macroweight.to_sql("dim_macroweight", conn2, index=False, if_exists="replace")
df_sectorweight.to_sql("dim_sectorweight", conn2, index=False, if_exists="replace")
```
